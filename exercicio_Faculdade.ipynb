```python
# exerccio 2 - Faa um programa que pergunte que dia , e diga quantos faltam at o final do ms. Considere o ms com trinta e um dia:
 
# Escrevendo a mensagem do sistema:

print("Estamos em um ms de 31 dias. deseja saber quantos dias falta ate o final do ms? ")
print("Responda com sim ou no!")
# escrevendo a variavel de entrada, chamada resposta, como um recurso de interatividade do usuario com o programa
# e utilizando para encerrar caso o usuario no queira
# Por enquanto apenas com condies.

resposta = input("")

# Processamento do cdigo com condies e operadores lgicos:

if resposta == "sim" or resposta == "Sim" or resposta == "SIM":
# colocando variavel para a continuidade do sistema.  
  dia = int(input("Diga-me o dia que estamos deste mes para saber quantos dias restam para acabar: "))
  # Escrevendo uma variavel para fazer o processamento.
  restante_do_mes = 31 - dia 
  # Trmino do programa
  print("Faltam", restante_do_mes,"dias para acabar este ms")
else: 
  # Trmino do programa
  print("Programa encerrado por opo do usurio!")
```

12. Elabore um algoritmo que leia um nmero de entrada que indicar a quantidade deregistros a serem lidos (N). Em seguida algoritmo deve solicitar o nome e idade de Npessoas e ao final apresentar o nome da pessoa mais velha.

```python
#Entrada de dados com uma varivel declarada
#varivel qtd = para receber o nmero que o usurio
#ir querer para a quantidade de vezes
qtd = int(input("Digite a quantidade de registros a serem lidos: "))

# Processamento dos dados
# Definindo listas
nomes = []
idades = []

#Processamento dos dados
#Lao de repetio: for
for i in range(qtd):
  #Declarando variveis dentro do lao para receber uma quantidade diferente de valores:
  #nome : recebe os nomes das pessoas para o registro
  #idade : recebe a idade das pessoas para o registro
    nome = input(f"Digite o nome da {i+1} pessoa: ")
    idade = int(input(f"Digite a idade de {nome}: "))
    #inserindo os valores atribuidos dentro das listas (com o mesmo nome das variavis)
    nomes.append(nome)
    idades.append(idade)

#processamento de dados
#Declarando variveis
indice_mais_velho = idades.index(max(idades))
nome_mais_velho = nomes[indice_mais_velho]

print(f"\nA pessoa mais velha  {nome_mais_velho}!")
```

13. Elabore um algoritmo que leia um nmero de entrada que indicar a quantidade deregistros a serem lidos (N). Em seguida algoritmo deve solicitar o sexo (M/F) e idade de Npessoas e ao final apresentar a mdia de idade de ambos os gneros catalogados

```python
# Leitura do nmero de registros
n = int(input("Digite a quantidade de registros a serem lidos: "))
# Inicializao das variveis de contagem
soma_idade_m = 0 # Soma das idades dos homens
soma_idade_f = 0 # Soma das idades das mulheres
qtd_m = 0 # Quantidade de homens
qtd_f = 0 # Quantidade de mulheres
# Loop para leitura dos registros
for i in range(n):
    # Leitura do sexo e idade da pessoa
    sexo = input("Digite o sexo da pessoa (M/F): ")
    idade = int(input("Digite a idade da pessoa: "))
    # Verificao do sexo da pessoa e atualizao das variveis de contagem
    if sexo == "M":
        soma_idade_m += idade
        qtd_m += 1
    elif sexo == "F":
        soma_idade_f += idade
        qtd_f += 1
# Clculo da mdia de idade de ambos os gneros catalogados
media_idade_m = soma_idade_m / qtd_m if qtd_m > 0 else 0
media_idade_f = soma_idade_f / qtd_f if qtd_f > 0 else 0
media_idade_total = (soma_idade_m + soma_idade_f) / n if n > 0 else 0
# Apresentao dos resultados
print("Mdia de idade dos homens catalogados: %.2f" % media_idade_m)
print("Mdia de idade das mulheres catalogadas: %.2f" % media_idade_f)
print("Mdia de idade total dos gneros catalogados: %.2f" % media_idade_total)
```

15. Elabore um algoritmo que solicite N nmeros reais e quando o usurio informar o valornulo 0 (zero) o programa ordene e mostre todos os nmeros informados de forma crescente

```python
print("Este programa tem como objetivo receber nmeros reais e lista-los em")
print("ordem crescente. Por favor, atenda ao pedido seguinte: ")
i = 0
reais = []
while True:
    num = float(input("Digite um nmero real ou 0 para sair do loop: "))
    i+=1
    reais.insert(i,num)
    reais.sort()
    if num == 0:
        break
print("Aqui esta seus nmeros reais listados em ordem crescente:")
print(reais)
```

16. Escreva um programa que v solicitando as idades dos alunos da sala at que todossejam informados (perguntar ao usurio se deseja informar a idade do prximo aluno). Aofinal apresentar a idade do mais novo, a idade do mais velho, Quantos alunos tm mais de18 anos, quantos alunos tm at 18 anos, a mdia aritmtica e a mediana.

```python
idades = []
continuar = 's'

while continuar.lower() == 's':
    idade = int(input("Digite a idade do aluno: "))
    idades.append(idade)
    continuar = input("Deseja informar a idade de outro aluno? (S/N) ")

# Clculo da idade do mais novo e do mais velho
idade_min = min(idades)
idade_max = max(idades)

# Clculo da quantidade de alunos com mais de 18 anos e at 18 anos
qtd_maior_18 = len([idade for idade in idades if idade > 18])
qtd_menor_ou_igual_18 = len([idade for idade in idades if idade <= 18])

# Clculo da mdia aritmtica das idades
media = sum(idades) / len(idades)

# Clculo da mediana das idades
idades_ordenadas = sorted(idades)
n = len(idades_ordenadas)
if n % 2 == 0:
    mediana = (idades_ordenadas[n // 2 - 1] + idades_ordenadas[n // 2]) / 2
else:
    mediana = idades_ordenadas[n // 2]

# Impresso dos resultados
print(f"Idade do aluno mais novo: {idade_min}")
print(f"Idade do aluno mais velho: {idade_max}")
print(f"Quantidade de alunos com mais de 18 anos: {qtd_maior_18}")
print(f"Quantidade de alunos com at 18 anos: {qtd_menor_ou_igual_18}")
print(f"Mdia aritmtica das idades: {media:.2f}")
print(f"Mediana das idades: {mediana}")
```
